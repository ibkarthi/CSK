```python
def cleancell(df):
    if 'Match tied' in df['Winner']:
        if df['Winner'] in df['Team1']:
            return df['Team1']
        else:
            return df['Team2']
    else:
        return df['Winner']
```

```python
iplt20['Winner'] = iplt20.apply(cleancell, axis = 1)
```

```python
def cleancell1(df):
    if 'Match tied' in df['Result']:
        if df['Result'] in df['Team1']:
            return (df['Team1'] + ' won by tie')
        else:
            return (df['Team2'] + ' won by tie')
    else:
        return df['Result']
```

```python
iplt20['Result'] = iplt20.apply(cleancell1, axis = 1)
```

```python
iplt20['Win by'] = iplt20['Result'].str.split(' won by ').str[1]
```

```python
iplt20[(iplt20['Win by'] == 'Super Over') | (iplt20['Win by'] == 'Boundary Count')]
```

```python
# Super Over - Punjab Kings(Chasing Team)
# Boundary Count - Rajasthan Royals(First batting Team)
def chase(df):
    if ('Runs' in df['Win by']) or ('Boundary Count' in df['Win by']):
        return 'Lost'
    elif 'tie' in df['Win by']:
        return 'Tie'
    else:
        return 'Won'
```

```python
iplt20['Chasing Team'] = iplt20.apply(chase, axis = 1)
```
