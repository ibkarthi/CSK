```python
totcsk = pd.read_sql_query(sql="select Match, Teams, Result from iplt20 where Match in (select Match from csk)",con=ipl_db)
totcsk.head()
```

### Finding the Chasing Team

```python
def chase(row):
    lst1 = row['Result'].split(' won by ')
    lst2 = row['Teams'].split(' v ')
    if 'Wicket' in lst1[-1]:
        return lst1[0]
    elif 'Run' in lst1[-1]:
        if lst2[0] != lst1[0]:
            return lst2[0]
        else:
            return lst2[1]
    else:
        return 'Doubt'
```

```python
totcsk['Chaser'] = totcsk.apply(chase, axis = 1)
```
